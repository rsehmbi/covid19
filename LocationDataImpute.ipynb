```python
import pandas as pd
import numpy as np
```

```python
ProcessedLocations = pd.read_csv("ImputedProvince.csv")
```

```python
ProcessedLocations.head()
ProcessedLocations.set_index('Unnamed: 0')
ProcessedLocations.columns
```

```python
ProcessedLocations = ProcessedLocations.dropna(how='any',axis=0) 
```

```python
ProcessedLocations['Province'] = ProcessedLocations["Province_State"]
ProcessedLocations.drop('Unnamed: 0', axis=1, inplace=True)
ProcessedLocations
```

```python
DataFrameNan = pd.read_csv("dataset/processed_location_Sep20th2020.csv")
```

```python
DataFrameNan = DataFrameNan[DataFrameNan['Province_State'].notna()]
DataFrameNan
```

```python
df3 = pd.concat([ProcessedLocations, DataFrameNan])
df3
```

```python
df3.drop(['geom', 'Province'], axis=1, inplace=True)
```

```python
df3.isnull().sum(axis = 0)
```

```python
LatLong = df3[df3['Lat'].isnull()]
LatLong = LatLong[~LatLong.Province_State.str.contains("Unknown")]
```

```python
LatLong.to_csv("LatLongMissing.csv")
```

```python
ImputedLat_Long = pd.read_csv("ImputedLongLat.csv")
```

```python
ImputedLat_Long = ImputedLat_Long[ImputedLat_Long['Lat'] != 0]
```

```python
df3 = df3[df3['Lat'].notna()]
```

```python
LongLatitudeImputation = pd.concat([df3, ImputedLat_Long])
```

```python
LongLatitudeImputation.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
```

```python
LongLatitudeImputation.isnull().sum(axis = 0)
```

```python
Incidence_Rate = LongLatitudeImputation[LongLatitudeImputation['Incidence_Rate'].isnull()]
```

```python
Incidence_Rate.to_csv("MissingIncidenceRate.csv")
```

```python
ImputedIncidenceRate = pd.read_csv("ImputedIncidenceRate.csv")
```

```python
IncidenceRatePresent = LongLatitudeImputation[LongLatitudeImputation['Incidence_Rate'].notna()]
```

```python
JoinedTable = pd.concat([IncidenceRatePresent, ImputedIncidenceRate])
JoinedTable.drop(['Unnamed: 0', 'Unnamed: 0.1','Population'], axis=1, inplace=True)
```

```python
JoinedTable.isnull().sum(axis = 0)
JoinedTable
```

```python
Ratio = JoinedTable[JoinedTable['Case-Fatality_Ratio'].isnull()]
CleanedData = JoinedTable[JoinedTable['Case-Fatality_Ratio'].notna()]
```

```python
Ratio['Case-Fatality_Ratio'] = Ratio['Case-Fatality_Ratio'].fillna(0)
```

```python
CleanedData.isnull().sum(axis = 0)
```

```python
Ratio.dropna(subset = ["Active"], inplace=True)
```

```python
FinalCleanedData = pd.concat([Ratio, CleanedData])
FinalCleanedData = FinalCleanedData.reset_index(drop=True)
FinalCleanedData.to_csv("dataset/CleanedLocations.csv")
```
